````python
import numpy as np

# ----- Define Fuzzy Sets -----
X = [1, 2, 3, 4, 5]
A = {1: 0.2, 2: 0.7, 3: 1.0, 4: 0.4, 5: 0.1}
B = {1: 0.6, 2: 0.2, 3: 0.9, 4: 0.3, 5: 0.5}

# ----- Fuzzy Set Operations -----
def union(A, B):
    return {x: max(A[x], B[x]) for x in A}

def intersection(A, B):
    return {x: min(A[x], B[x]) for x in A}

def complement(A):
    return {x: 1 - A[x] for x in A}

print("Union:", union(A,B))
print("Intersection:", intersection(A,B))
print("Complement of A:", complement(A))

# ----- Cartesian Product (Fuzzy Relation) -----
def cartesian(A, B):
    return {(x, y): min(A[x], B[y]) for x in A for y in B}

R1 = cartesian(A, B)
R2 = cartesian(B, A)  # another relation

print("\nFuzzy Relation R1 (A x B):")
for pair, val in R1.items():
    print(pair, ":", val)
print("\nFuzzy Relation R2 (B x A):")
for pair, val in R2.items():
    print(pair, ":", val)

# ----- Max-Min Composition -----
def max_min_composition(R1, R2, A_set, B_set, C_set):
    result = {}
    for x in A_set:
        for z in C_set:
            values = []
            for y in B_set:
                values.append(min(R1[(x,y)], R2[(y,z)]))
            result[(x,z)] = max(values)
    return result

Comp = max_min_composition(R1, R2, X, X, X)

print("\nMax-Min Composition (R1 ‚àò R2):")
for pair, val in Comp.items():
    print(pair, ":", val)













'''
Perfect üëç ‚Äî let‚Äôs make this simple so you can **explain it clearly to your ma‚Äôam**.
Here‚Äôs exactly **how you should describe the code and what each output means** step by step üëá

---

## ü™Ñ **How to Explain the Code to Your Ma‚Äôam**

### **1Ô∏è‚É£ Start with What the Code Does Overall**

> ‚ÄúThis code demonstrates basic **fuzzy logic operations** ‚Äî like union, intersection, complement, and fuzzy relations ‚Äî using Python.
> It also shows how to combine fuzzy relations using **max‚Äìmin composition**.‚Äù

---

### **2Ô∏è‚É£ Explain What Fuzzy Sets Are**

> ‚ÄúIn fuzzy sets, elements don‚Äôt just belong or not belong.
> Each element has a **membership value between 0 and 1**, showing how strongly it belongs to the set.‚Äù

Example:

* 1 means *fully belongs*
* 0 means *does not belong*
* Between 0 and 1 means *partially belongs*

---

### **3Ô∏è‚É£ Explain the Input**

```python
X = [1, 2, 3, 4, 5]
A = {1: 0.2, 2: 0.7, 3: 1.0, 4: 0.4, 5: 0.1}
B = {1: 0.6, 2: 0.2, 3: 0.9, 4: 0.3, 5: 0.5}
```

> ‚ÄúHere, X is the universe (the list of all elements).
> A and B are fuzzy sets that assign each element in X a degree of belonging.‚Äù

---

### **4Ô∏è‚É£ Explain Each Operation and Its Output**

#### ‚úÖ **Union**

```python
Union: {1:0.6, 2:0.7, 3:1.0, 4:0.4, 5:0.5}
```

> ‚ÄúUnion means combining both sets ‚Äî we take the **maximum membership** of each element.
> So the result shows how strongly each element belongs to A or B.‚Äù

Example:
Element 3 = max(1.0, 0.9) = 1.0 ‚Üí fully belongs to the union.

---

#### ‚úÖ **Intersection**

```python
Intersection: {1:0.2, 2:0.2, 3:0.9, 4:0.3, 5:0.1}
```

> ‚ÄúIntersection means common part ‚Äî we take the **minimum membership** of each element.
> It shows how strongly each element belongs to both sets at the same time.‚Äù

Example:
Element 3 = min(1.0, 0.9) = 0.9 ‚Üí high overlap.

---

#### ‚úÖ **Complement**

```python
Complement of A: {1:0.8, 2:0.3, 3:0.0, 4:0.6, 5:0.9}
```

> ‚ÄúComplement means opposite of the set.
> We subtract each value from 1 ‚Äî so elements that were strongly in A now become weakly in the complement.‚Äù

Example:
Element 3 = 1 ‚àí 1.0 = 0 ‚Üí not in complement.
Element 5 = 1 ‚àí 0.1 = 0.9 ‚Üí almost fully in complement.

---

### **5Ô∏è‚É£ Fuzzy Relation (Cartesian Product)**

```python
R1 = A √ó B
```

> ‚ÄúA fuzzy relation connects elements from A and B.
> Each pair (x,y) has a membership equal to **min(A[x], B[y])**.‚Äù

So you‚Äôll see output like:

```
(1,1): 0.2
(1,2): 0.2
(2,3): 0.7
(3,3): 0.9
...
```

> ‚ÄúThis means element 1 from A and 1 from B are related with degree 0.2,
> and 3 from A and 3 from B are strongly related with degree 0.9.‚Äù

Same idea for R2 = B √ó A (just reversed).

---

### **6Ô∏è‚É£ Max‚ÄìMin Composition**

```python
Max-Min Composition (R1 ‚àò R2):
(1,1): ...
(1,2): ...
...
```

> ‚ÄúThis part combines two fuzzy relations, R1 (A√óB) and R2 (B√óA).
> It calculates how strongly each element in A relates to each element in A *through* B.‚Äù

How it works (say for (x=1, z=3)):

* For each y in B, take `min(R1(x,y), R2(y,z))`
* Then take the **maximum** of all those mins

> ‚ÄúThis gives the final relation (A to A) based on indirect connections through B.‚Äù

So the printed output shows the **strength of indirect relationships**.

---

### **7Ô∏è‚É£ Wrap-Up Line (to Impress üòé)**

> ‚ÄúIn summary, the code demonstrates how fuzzy logic handles **uncertainty** and **partial truth**.
> It calculates how sets and relations interact when elements have degrees of belonging instead of being simply true or false.‚Äù

---

## üó£Ô∏è Example of How You Can Say It (short version for viva)

> ‚ÄúThis Python code performs basic fuzzy set operations like union, intersection, and complement using membership values.
> Then it forms fuzzy relations (A√óB and B√óA) using the Cartesian product, and finally performs a **max‚Äìmin composition** to combine two relations.
> The outputs show how strongly elements are connected ‚Äî for example, in the union we take the max of memberships, in intersection the min, and in composition the strongest indirect relationship.‚Äù



'''
````
