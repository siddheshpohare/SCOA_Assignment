````python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Step 1: Load stock data
ticker = 'AAPL'
df = yf.download(ticker, start='2020-01-01', end='2023-01-01')
df['Target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)

# Step 2: Feature selection and scaling
features = df[['Open', 'High', 'Low', 'Close', 'Volume']]
scaler = MinMaxScaler()
X = scaler.fit_transform(features)
y = df['Target'].values

# Step 3: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 4: Build ANN model
model = Sequential()
model.add(Dense(64, input_dim=5, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)

# Step 5: Evaluate model
y_pred = (model.predict(X_test) > 0.5).astype("int32")
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)









'''
Perfect ğŸ‘ Letâ€™s go through this **step-by-step explanation** of your stock prediction ANN code â€”
so that you can easily explain it to your teacher (â€œmaâ€™amâ€) clearly and confidently.

---

## ğŸ§  **Goal of the Code**

The aim of this code is to **predict whether Appleâ€™s (AAPL) stock price will go up or down the next day**, using **an Artificial Neural Network (ANN)** model trained on past stock data.

---

## ğŸ§© **Step-by-Step Explanation**

---

### **Step 1: Importing Libraries**

```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

âœ… **Explanation:**

* **NumPy & Pandas:** Handle numerical and tabular data.
* **yfinance:** Fetches stock price data directly from Yahoo Finance.
* **MinMaxScaler:** Scales numeric data between 0 and 1 to help the neural network learn faster.
* **train_test_split:** Splits data into training and testing sets.
* **accuracy_score & confusion_matrix:** Used for performance evaluation.
* **Keras (TensorFlow):** Used to build and train the Artificial Neural Network.

---

### **Step 2: Load Stock Data**

```python
ticker = 'AAPL'
df = yf.download(ticker, start='2020-01-01', end='2023-01-01')
df['Target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

âœ… **Explanation:**

* This downloads **Apple Inc. (AAPL)** stock data from **2020 to 2023**.
* It includes columns like: `Open`, `High`, `Low`, `Close`, `Volume`.
* We create a new column called **`Target`**:

  * **1** â†’ Tomorrowâ€™s closing price is **higher** than todayâ€™s (price **went up**)
  * **0** â†’ Tomorrowâ€™s closing price is **lower or same** (price **didnâ€™t go up**)

ğŸ’¡ **This â€œTargetâ€ column is our prediction goal.**

---

### **Step 3: Select Features and Scale Them**

```python
features = df[['Open', 'High', 'Low', 'Close', 'Volume']]
scaler = MinMaxScaler()
X = scaler.fit_transform(features)
y = df['Target'].values
```

âœ… **Explanation:**

* The model uses **five input features** (`Open`, `High`, `Low`, `Close`, `Volume`).
* These features are **scaled between 0 and 1** â€” this ensures all values are on a similar range and prevents large numbers (like â€œVolumeâ€) from dominating.
* `X` â†’ input data
  `y` â†’ output labels (0 or 1)

---

### **Step 4: Split into Training and Testing Data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

âœ… **Explanation:**

* We use **80%** of the data for **training** and **20%** for **testing**.
* `shuffle=False` means the data is **not randomly shuffled** â€” since stock data is time-based, we keep the sequence intact (training on past â†’ testing on future).

---

### **Step 5: Build the Artificial Neural Network**

```python
model = Sequential()
model.add(Dense(64, input_dim=5, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

âœ… **Explanation:**
This builds a **3-layer feedforward neural network**:

| Layer | Type  | Neurons | Activation | Purpose                          |
| ----- | ----- | ------- | ---------- | -------------------------------- |
| 1     | Dense | 64      | ReLU       | Learn basic patterns from inputs |
| 2     | Dense | 32      | ReLU       | Learn deeper relations           |
| 3     | Dense | 1       | Sigmoid    | Output probability (0 to 1)      |

* **ReLU (Rectified Linear Unit):** Helps network learn efficiently by avoiding vanishing gradient.
* **Sigmoid:** Converts final output to a probability between 0 and 1 (helps classify up/down).

---

### **Step 6: Compile and Train the Model**

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)
```

âœ… **Explanation:**

* **Loss function:** `binary_crossentropy` â€“ good for binary classification (0 or 1).
* **Optimizer:** `adam` â€“ adjusts learning rates automatically for faster convergence.
* **epochs=50:** Trains the model for 50 complete passes through the dataset.
* **batch_size=32:** Updates weights after every 32 samples.
* The training process shows progress per epoch (loss and accuracy improving gradually).

---

### **Step 7: Evaluate Model Performance**

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```

âœ… **Explanation:**

* The model predicts probabilities (like 0.7 = 70% chance stock will go up).
* `(>0.5)` converts probability into a binary class:

  * > 0.5 â†’ 1 (price goes up)
  * â‰¤0.5 â†’ 0 (price goes down)
* We calculate **accuracy** and **confusion matrix** to evaluate performance.

---

## ğŸ§¾ **Understanding the Output**

### **1ï¸âƒ£ Accuracy:**

Example output:

```
Accuracy: 0.62
```

ğŸ‘‰ This means the model correctly predicted the direction of stock movement **62% of the time** on test data.

### **2ï¸âƒ£ Confusion Matrix:**

Example:

```
[[95 40]
 [55 110]]
```

|                     | Predicted Down (0) | Predicted Up (1) |
| ------------------- | ------------------ | ---------------- |
| **Actual Down (0)** | 95                 | 40               |
| **Actual Up (1)**   | 55                 | 110              |

âœ… **Interpretation:**

* **95** â†’ Correctly predicted â€œDownâ€
* **110** â†’ Correctly predicted â€œUpâ€
* **40 + 55 = 95 mistakes** â†’ Wrong predictions

The model is **moderately accurate**, but can be improved with:

* More technical indicators (like moving averages, RSI, etc.)
* Longer training duration
* LSTM (for time-series data)

---

## ğŸ§  **Summary (for easy oral explanation)**

You can say:

> â€œThis code predicts whether Appleâ€™s stock will go up or down tomorrow using a neural network.
> It first downloads the stock data using yfinance, then creates a target column showing if the price increased the next day.
> We scale the input features and train an ANN with two hidden layers using ReLU and a sigmoid output layer for binary prediction.
> Finally, we evaluate accuracy and display a confusion matrix to check how many times the model predicted correctly.â€

---


'''
````
